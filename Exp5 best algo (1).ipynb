```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#importing dataset
import pandas as pd
import io
df=pd.read_csv('Regression.csv')
df
```

```python
#RegressionFor simple linear regression we need one input and one output.. so we are taking Age as input and estimated salary as output from this dataset
x1=df.iloc[:,[2]]
x1
```

```python
y1=df['EstimatedSalary'].values
y1
```

```python
#building the model
model1=LinearRegression().fit(x1,y1)
r_sq=model1.score(x1,y1)
```

```python
#evaluating the model
print("coefficient of determination:",r_sq)
print("Intercept:",model1.intercept_)
print("Slope:",model1.coef_)
```

```python
'''Multiple Linear Regression
In multiple linear regression we have multiple inputs and one output.. so inputs are age,purchased,gender and estimated salary as output
since gender is in string formatwe are converting it into integer format using dummies'''
```

```python
#data preprocessing (data transfromation)
ed=pd.get_dummies(df,columns=['Gender'])
print(ed)
```

```python
ed.Gender_Female.replace('False',0,inplace=True)
#ed.Gender_Female.replace('True',1,inplace=True)
#ed.Gender_Male.replace('False',0,inplace=True)
#ed.Gender_Male.replace('True',1,inplace=True)
ed
```

```python
x2=ed.iloc[:,[1,3,4,5]]
x2
```

```python
y2=ed['EstimatedSalary'].values
y2
```

```python
#building the model
model2=LinearRegression().fit(x2,y2)
r_sq=model2.score(x2,y2)
```

```python
#evaluating the model
print("Coefficient of determination:",r_sq)
print("Intercept:",model2.intercept_)
print("Slope:",model2.coef_)
```

```python
'''Polynomial Regression
For polynomial regression we are taking same input and output from simple linear regression'''
```

```python
#building the model
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=3)
x1_poly=poly_reg.fit_transform(x1)
model3=LinearRegression().fit(x1_poly,y1)
r_sq=model3.score(x1_poly,y1)
```

```python
#evaluating the model
print("Coefficient of determination:",r_sq)
print("Intercept:",model3.intercept_)
print("Slope:",model3.coef_)
```
